```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib.patches import Ellipse
```

```python
# Define custom colors
new_red = '#FF6384'      # Reddish-pink, coral-like
new_blue = '#36A2EB'     # Vibrant medium blue
new_teal = '#4BC0C0'     # Cyan-green teal
new_green = '#2ECC71'    # Bright emerald green
new_purple = '#9B59B6'   # Vibrant purple
new_orange = '#E67E22'   # Warm orange
```

```python
eb_no_db_range = np.linspace(0, 20, 20)
print(eb_no_db_range)
```

```python
num_bits_per_symbol = 4
eb_no_db_range = np.linspace(0, 20, 20)

mpl.rcParams['font.family'] = 'serif'  # or 'sans-serif', 'monospace', etc.
mpl.rcParams['font.serif'] = ['Times New Roman']  # or any available font on your system
mpl.rcParams['font.size'] = 14  # Increase this for larger text

ser_ae_16QAM_4 = np.loadtxt("ser_ae_16QAM_4.csv", delimiter = ',')
ser_ae_16QAM_8 = np.loadtxt("ser_ae_16QAM_8.csv", delimiter = ',')
ser_ae_16QAM_12 = np.loadtxt("ser_ae_16QAM_12.csv", delimiter = ',')

ser_ae_64QAM_4 = np.loadtxt("ser_ae_64QAM_4.csv", delimiter = ',')
ser_ae_64QAM_8 = np.loadtxt("ser_ae_64QAM_8.csv", delimiter = ',')
ser_ae_64QAM_12 = np.loadtxt("ser_ae_64QAM_12.csv", delimiter = ',')

ser_ae_256QAM_4 = np.loadtxt("ser_ae_256QAM_4.csv", delimiter = ',')
ser_ae_256QAM_8 = np.loadtxt("ser_ae_256QAM_8.csv", delimiter = ',')
ser_ae_256QAM_12 = np.loadtxt("ser_ae_256QAM_12.csv", delimiter = ',')

plt.figure(figsize=(9, 3))
plt.semilogy(eb_no_db_range, ser_ae_16QAM_4, 'o-', color=new_blue, markersize=3, linewidth=2, label=f'AE-based E2E {2**num_bits_per_symbol}-QAM, 4dB')
plt.semilogy(eb_no_db_range, ser_ae_16QAM_8, 'o-', color=new_orange, markersize=3, linewidth=2, label=f'AE-based E2E {2**num_bits_per_symbol}-QAM, 8dB')
plt.semilogy(eb_no_db_range, ser_ae_16QAM_12, 'o-', color=new_green, markersize=3, linewidth=2, label=f'AE-based E2E {2**num_bits_per_symbol}-QAM, 12dB')

plt.semilogy(eb_no_db_range, ser_ae_64QAM_4, '--', color=new_blue, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+2)}-QAM, 4dB')
plt.semilogy(eb_no_db_range, ser_ae_64QAM_8, '--', color=new_orange, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+2)}-QAM, 8dB')
plt.semilogy(eb_no_db_range, ser_ae_64QAM_12, '--', color=new_green, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+2)}-QAM, 12dB')

plt.semilogy(eb_no_db_range, ser_ae_256QAM_4, '-', color=new_blue, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+4)}-QAM, 4dB')
plt.semilogy(eb_no_db_range, ser_ae_256QAM_8, '-', color=new_orange, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+4)}-QAM, 8dB')
plt.semilogy(eb_no_db_range, ser_ae_256QAM_12, '-', color=new_green, markersize=3, linewidth=2, label=f'AE-based E2E {2**(num_bits_per_symbol+4)}-QAM, 12dB')


annotation_index_1 = 10
ellipse_1 = Ellipse(xy=(eb_no_db_range[annotation_index_1]+0.2, ser_ae_16QAM_8[annotation_index_1]),
                    width=1, height=0.001,
                    facecolor='none', edgecolor='red', linewidth=2)
plt.gca().add_patch(ellipse_1)
plt.annotate('M = 16', 
             xy=(eb_no_db_range[annotation_index_1], ser_ae_16QAM_8[annotation_index_1]), 
             xytext=(eb_no_db_range[annotation_index_1]-2 , ser_ae_16QAM_8[annotation_index_1]-0.001),
             fontsize=12, color='red')



annotation_index_2 = 15
ellipse_2 = Ellipse(xy=(eb_no_db_range[annotation_index_2]-1.25, ser_ae_64QAM_4[annotation_index_2]+0.00045),
                    width=3, height=0.005,
                    facecolor='none', edgecolor='red', linewidth=2)
plt.gca().add_patch(ellipse_2)
plt.annotate('M = 64',
             xy=(eb_no_db_range[annotation_index_2], ser_ae_64QAM_4[annotation_index_2]),
             xytext=(eb_no_db_range[annotation_index_2]-1 , ser_ae_64QAM_4[annotation_index_2]+0.02),
             fontsize=12, color='red')


annotation_index_3 = 12
ellipse_3 = Ellipse(xy=(eb_no_db_range[annotation_index_3]+0.5, ser_ae_256QAM_8[annotation_index_3]),
                    width=1, height=0.2,
                    facecolor='none', edgecolor='red', linewidth=2)
plt.gca().add_patch(ellipse_3)
plt.annotate('M = 256',
             xy=(eb_no_db_range[annotation_index_3], ser_ae_256QAM_8[annotation_index_3]-0.1),
             xytext=(eb_no_db_range[annotation_index_3]+1 , ser_ae_256QAM_8[annotation_index_3]+0.1),
             fontsize=12, color='red')

plt.xlabel('Eb/N0 (dB)', fontsize=15)
# plt.title(f'Symbol-wise AI-based Transceivers AE-CNN: Comparision SER under AWGN channel', fontsize= 10)
plt.legend(loc='lower left', fontsize=8, frameon=True, shadow=True, fancybox=True)
plt.xlabel('Eb/N0 (dB)', fontsize=15)
plt.ylabel('Symbol Error Rate (SER)', fontsize=15)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.6)

plt.ylim(1e-4, 2)
plt.xlim(0,20)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)



plt.show()
```
